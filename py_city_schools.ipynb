```python
#import modules
import pandas as pd
import numpy as np
```

```python
#load data

# File to Load
school_data= "schools_complete.csv"
student_data = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data)
student_data = pd.read_csv(student_data)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary:

```python
#columns
school_data_complete.columns
```

```python
#total number of schools
schools_num=len(np.unique(school_data_complete['school_name']))

#total number of students
num_stu=school_data_complete.shape[0]

#total budget
budget_tot=school_data['budget'].sum()

#average math score
avg_math=round(school_data_complete['math_score'].mean(),2)

#average reading score
avg_reading=round(school_data_complete['reading_score'].mean(),2)

#overall passing rate
overall_perc=round((avg_math+avg_reading)/2,2)

#math pass percentage
pass_math_perc=round(school_data_complete[school_data_complete['math_score'] >= 70].shape[0]*100/num_stu,4)

#reading pass percentage
pass_reading_perc=round(school_data_complete[school_data_complete['reading_score'] >= 70].shape[0]*100/num_stu,4)

#put data into pandas data frame table
district_summary = pd.DataFrame([{
    'Total Schools': schools_num,'Total Students': num_stu,'Total Budget': f"${budget_tot}",
    'Average Math Score': f"{avg_math}%",'Average Reading Score': f"{avg_reading}%",
    '% Passing Math': f"{round(pass_math_perc,2)}%",'% Passing Reading': f"{round(pass_reading_perc,2)}%",
    '% Overall Passing': overall_perc
}])

#display
district_summary
```

## School Summary

```python
#School Name/School Type
school_summary = pd.DataFrame(list(school_data_complete.groupby(['school_name','type','size','budget']).groups))
school_summary.columns=['School Names','Type','Total Students','School Budget']

#per student budget
school_summary['Per Student Budget'] = school_summary['School Budget']/school_summary['Total Students']

#avg math score
school_summary['Avg Math Score']=pd.DataFrame(school_data_complete.groupby('school_name')['math_score'].mean().values)

#avg reading score
school_summary['Avg Reading Score']=pd.DataFrame(school_data_complete.groupby('school_name')['reading_score'].mean().values)

#percent passing math
passed = school_data_complete[school_data_complete['math_score'] >= 70]
students_passed = passed.groupby('school_name')['math_score'].count().values
school_summary['% Passing Math'] = students_passed/school_summary['Total Students']

#percent passing reading
passed = school_data_complete[school_data_complete['reading_score'] >= 70]
students_passed = passed.groupby('school_name')['reading_score'].count().values
school_summary['% Passing Reading'] = students_passed/school_summary['Total Students']

#overall passing rate
school_summary['Overall Passing Rate'] = (school_summary['% Passing Math'] + school_summary['% Passing Reading'])/2

#output table
school_summary
```

## Top Performing Schools (By Passing Rate)

```python
#initialize new data frame with school summary table sorted by passing rate
passing_rate_sort =school_summary.sort_values(by=['Overall Passing Rate'],ascending=False).iloc[0:5,:].reset_index(drop=True)

#format percentages to integers
passing_rate_sort['Overall Passing Rate'] = passing_rate_sort['Overall Passing Rate']*100
passing_rate_sort['% Passing Math'] = passing_rate_sort['% Passing Math']*100
passing_rate_sort['% Passing Reading'] = passing_rate_sort['% Passing Reading']*100

#display table
passing_rate_sort
```

##  Bottom Performing Schools (By Passing Rate)

```python
#initialize new data frame with first 5 rows of school summary table sorted by passing rate
bottom_rate_sort = school_summary.sort_values(by=['Overall Passing Rate'],ascending=True).iloc[0:5,:].reset_index(drop=True)

#format percentages
bottom_rate_sort['Overall Passing Rate'] = bottom_rate_sort['Overall Passing Rate']*100
bottom_rate_sort['% Passing Math'] = bottom_rate_sort['% Passing Math']*100
bottom_rate_sort['% Passing Reading'] = bottom_rate_sort['% Passing Reading']*100

#display table
bottom_rate_sort
```

## Math Scores by Grade

```python
#function that outputs mean subject scores by grade
#inputs: school_data_complete merged data set/list of grades interested in/subject: 'math' or 'reading'
def scores_by_grade(school_data_complete, grade_list, subject):
    
    #set the actual str column name we will sort by later
    if subject == 'math':
        score = 'math_score'
    else:
        score = 'reading_score'
    
    #initialize list
    grade=[]
    
    #for loop to fill averages of all grades for each school
    for i in range(len(grade_list)):
        x = school_data_complete[school_data_complete['grade'] == grade_list[i]]
        x_subject = x.groupby('school_name')[score].mean().values
        grade.append(x_subject)
    
    #names of the schools
    schools = pd.DataFrame(list(school_data_complete.groupby(['school_name']).groups))
    
    #convert to data frame
    grade=pd.DataFrame(grade)
    #transpose
    grade=np.transpose(grade)
    #name columns
    grade.columns=grade_list
    #insert new column with names of the schools
    grade.insert(loc=0, value=schools, column='School')
    #return value
    return(grade)
```

```python
#list of grades interested in for the function
grade_list = ['9th', '10th', '11th', '12th']
```

```python
scores_by_grade(school_data_complete, grade_list, 'math')
```

## Reading Scores by Grade

```python
#function change to subject 'reading'
scores_by_grade(school_data_complete, grade_list, 'reading')
```

## Scores by School Spending

```python
#because we will be solving for the same variables for different categories/bins, this is a function that can calculate
#required variables with category of interest
#INPUT: merged table: binned/cut table ---- group_names: names of each group ---- category: 'spending'/'size'/'type'
def scores_by_category(merged_table,group_names,category):
    
    if category == 'spending':
        cat='Per Student Budget'
    else:
        if category == 'size':
            cat='Total Students'
        else:
            cat='type'
    
    #create new data frame which will be our final output
    #values get mixed up on the 'type' option. Hard code bug fix
    if cat=='type':
        scores_by_school = pd.DataFrame([group_names[1],group_names[0]])
    else:
        scores_by_school = pd.DataFrame(group_names)
        
    #label 1st column    
    scores_by_school.columns = ['Bin']

    #average math score for each budget bin
    avg_math_score = pd.DataFrame(merge_table.groupby(cat)['math_score'].mean().values)
    #add to table
    scores_by_school['Average Math Scores'] = avg_math_score

    #average reading score for each budget bin
    avg_reading_score = pd.DataFrame(merge_table.groupby(cat)['reading_score'].mean().values)
    #add to table
    scores_by_school['Average Reading Scores'] = avg_reading_score

    #average percent passing math
    passing_group_m = merge_table[merge_table['math_score'] >= 70]

    #number of students that passed math in each bin
    math_pass_p = passing_group_m.groupby(cat)['math_score'].count().values

    #get total students in each bin
    bins_total_stu = [len(merge_table[merge_table[cat]==group_names[i]]) for i in range(len(group_names))]
    #values get mixed up on the 'type' option. Hard code bug fix
    if cat=='type':
        bins_total_stu[0], bins_total_stu[1] = bins_total_stu[1], bins_total_stu[0]
    
    #percent that pass math for each bin
    math_bin_pp = pd.DataFrame(math_pass_p/bins_total_stu)

    #add to table
    scores_by_school['% Passing Math'] = math_bin_pp*100

    #average reading score for each budget bin
    passing_group_r = merge_table[merge_table['reading_score'] >= 70]

    #number of students that passed math in each bin
    reading_pass_p = passing_group_r.groupby(cat)['reading_score'].count().values

    #percent that pass math for each bin
    reading_bin_pp = pd.DataFrame(reading_pass_p/bins_total_stu)
    
    #add to table
    scores_by_school['% Passing Reading'] = reading_bin_pp*100

    total_pass_rate = (math_bin_pp + reading_bin_pp)/2*100

    #add to table
    scores_by_school['% Overall Passing Rate'] = total_pass_rate

    #output final table
    return(scores_by_school)
```

```python
#we need some columns from our school summary report from earlier
school_extra_data = school_summary[['School Names','Per Student Budget']]

#change the name of the column so they have the same name for the merge
school_extra_data.columns = ['school_name','Per Student Budget']

#merge data sets
merge_table = pd.merge(school_data_complete, school_extra_data, on="school_name")
```

```python
# bins
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

#apply bin labels to data set
merge_table["Per Student Budget"] = pd.cut(merge_table["Per Student Budget"], spending_bins, labels=group_names)
```

```python
scores_by_category(merge_table,group_names,'spending')
```
